```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Dataset artwork cleaning

```python
museum_df = pd.read_csv('/Users/gabrielasena/Desktop/final-project-spiced/data/artworks.csv')
```

```python
museum_df.head(10)
```

```python
museum_df.isna().sum()
```

```python
museum_df.drop(['Diameter (cm)', 
                'Circumference (cm)', 
                'Height (cm)', 
                'Length (cm)',
                'Width (cm)',
                'Depth (cm)',
                'Weight (kg)',
                'Duration (s)'
               ], 
               axis=1, 
               inplace=True)
```

```python
museum_df.dropna(inplace= True)
```

```python
museum_df.info()
```

```python
museum_df.columns = map(str.lower, museum_df.columns)
```

```python
museum_df.rename(columns = {'date':'year', 
                            'name':'name_artist',
                            'artwork id': 'artwork_id',
                            'artist id':'artist_id',
                            'object number':'object_number',
                            'acquisition date':'acquisition_date'}, 
                     inplace= True)
```

```python
museum_df['acquisition_date'] = pd.to_datetime(museum_df['acquisition_date'])
```

```python
museum_df.reset_index(drop = True, inplace= True)
```

```python
museum_df.info()
```

```python
museum_df.head(5)
```

```python
museum_df.to_csv('/Users/gabrielasena/Desktop/final-project-spiced/data/artworks_cleaned.csv', index=False)
```

# Dataset artists cleaning

```python
artists_df = pd.read_csv('/Users/gabrielasena/Desktop/final-project-spiced/data/artists.csv')
```

```python
artists_df.head(10)
```

```python
artists_df.isna().sum()
```

```python
artists_df.columns = map(str.lower, artists_df.columns)
```

```python
artists_df.rename(columns = {'artist id': 'artist_id',
                             'birth year': 'birth_year',
                             'death year': 'death_year'}, 
                     inplace= True)
```

```python
artists_df.head(10)
```

```python
artists_df.info()
```

```python
artists_df['birth_year'] = artists_df['birth_year'].fillna(0.0).astype(int)
```

```python
artists_df['death_year'] = artists_df['death_year'].fillna(0.0).astype(int)
```

```python
artists_df['artist_id']= artists_df['artist_id'].map(str)
```

```python
artists_df.info()
```

```python
artists_df.head(10)
```

```python
artists_df.to_csv('/Users/gabrielasena/Desktop/final-project-spiced/data/artists_cleaned.csv', index=False)
```
